```python
import pandas as pd
```

```python
data = pd.read_csv('IDA.csv')
print(data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
battery_data = data[['Brand', 'Battery (mAh)']]
plt.figure(figsize=(10, 6))
plt.bar(battery_data['Brand'], battery_data['Battery (mAh)'], color='skyblue')

# Adding title and labels
plt.title('Battery Features', fontsize=16)
plt.xlabel('Mobile Phone Brands', fontsize=12)
plt.ylabel('Battery Power (mAh)', fontsize=12)

# Display the plot
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data['Camera (MP)'] = data['Camera (MP)'].str.split('+').str[0].astype(int)

# Extracting the necessary columns
model_data = data[['Model', 'Storage (GB)', 'Camera (MP)']]

# Create a multi-line plot
plt.figure(figsize=(10, 6))

# Plotting Storage and Camera against Model
plt.plot(model_data['Model'], model_data['Storage (GB)'], label='Storage (GB)', marker='o')
plt.plot(model_data['Model'], model_data['Camera (MP)'], label='Camera (MP)', marker='s')

# Adding title and labels
plt.title('Models vs Storage and Camera Features', fontsize=16)
plt.xlabel('Mobile Phone Models', fontsize=12)
plt.ylabel('Features', fontsize=12)

# Adding legend to differentiate between the lines
plt.legend()

# Display the plot
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
data['Camera (MP)'] = data['Camera (MP)'].str.split('+').str[0].astype(int)

# Extracting the necessary columns
model_data = data[['Model', 'Storage (GB)', 'Camera (MP)', 'Brand']]

# Create a multi-line plot
plt.figure(figsize=(10, 6))

# Plotting Storage and Camera against Model
plt.plot(model_data['Model'], model_data['Storage (GB)'], label='Storage (GB)', marker='o')
plt.plot(model_data['Model'], model_data['Camera (MP)'], label='Camera (MP)', marker='s')

# Adding title and labels
plt.title('Models vs Storage and Camera Features', fontsize=16)
plt.xlabel('Mobile Phone Brands', fontsize=12)
plt.ylabel('Features', fontsize=12)

# Customizing x-axis ticks with Brand names
plt.xticks(ticks=range(len(model_data['Model'])), labels=model_data['Brand'], rotation=45, ha='right')

# Adding legend to differentiate between the lines
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
